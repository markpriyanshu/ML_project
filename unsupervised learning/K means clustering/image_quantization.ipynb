```python
import matplotlib.image as mpimg
```

```python
image_as_array = mpimg.imread('palm_trees.jpg')
```

```python
image_as_array.shape # (1401, 934, 3) ---> (height ,width ,color) ---> (h,w,c)  , c= red ,green ,blue (rgb)
```

```python
(h,w,c) = image_as_array.shape
```

```python
h
```

```python
# 3D to 2D matrix conversion
# rows = h*w and columns =c
image_as_array_2D = image_as_array.reshape(h*w ,c) 
```

```python
image_as_array_2D.shape
```

```python
image_as_array_2D
```

```python
h*w # therefor total no. of row are = 1308534
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
label = model.fit_predict(image_as_array_2D)
```

```python
label
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
quantized_image = np.reshape(rgb_codes[label] , (h,w,c))
```

```python
plt.imshow(quantized_image)
```
