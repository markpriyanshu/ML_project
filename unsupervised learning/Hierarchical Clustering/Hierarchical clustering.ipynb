```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('cluster_mpg.csv')
```

```python
df
```

```python
df['origin'].unique()
```

```python
df.describe()
```

```python
df_with_dummies = pd.get_dummies(df.drop('name' , axis = 1))
```

```python
df_with_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
```

```python
scaled_X = scaler.fit_transform(df_with_dummies)
```

```python
scaled_X
```

```python
df_with_dummies.columns
```

```python
scaled_df = pd.DataFrame(data=scaled_X , columns=df_with_dummies.columns)
```

```python
scaled_df
```

```python
sns.heatmap(scaled_df.corr())
```

```python
sns.clustermap(scaled_df ,col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize = (12,4),dpi=200)
sns.scatterplot(data =df , x = 'mpg' , y = 'weight' ,hue = cluster_labels)
```

```python
plt.figure(figsize = (12,4),dpi=200)
sns.scatterplot(data =df , x = 'mpg' , y = 'horsepower' ,hue = cluster_labels)
```
