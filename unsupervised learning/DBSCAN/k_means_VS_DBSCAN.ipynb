## define a function which takes two arguments as model and data

```python
def display_model(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data = data , x = 'X1' , y = 'X2' , hue = labels , palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)
```

```python
display_model(km , df_moon)
```

```python
display_model(km , df_blob)
```

```python
display_model(km , df_circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=1)
```

```python
display_model(dbscan,df_blob)
# red points showing outliers
```

```python
dbscan_low = DBSCAN(eps=0.2)
```

```python
display_model(dbscan_low,df_moon)
```

```python
display_model(dbscan_low,df_circles)
```
