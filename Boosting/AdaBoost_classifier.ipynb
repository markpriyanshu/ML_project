```python
X = df.drop('class' , axis=1)
```

```python
X= pd.get_dummies(X , drop_first=True)
```

```python
X
```

```python
y = df['class']
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
'''n_estimators = 1 ; means we are allowing to make only one stump which is based on the most important features
(i.e. it has maximum gain or least impurity) from all the features  '''
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
model.feature_importances_
```

As in above array we can see that the most important feature is given the value 1 bcoz we are not considering only one feature.

```python
model.feature_importances_.argmax()# it will show the index no. having maximum value
```

```python
from sklearn.metrics import classification_report,accuracy_score,plot_confusion_matrix
```

```python
print(classification_report(y_test,preds))
```

#### From this we can say that by considering only odor we can classify maximum of mushrooms as poisnous and edible

### Plot a curve between n_estimators vs errors

```python
error_rate = []
for n in range(1,96):
    model_n = AdaBoostClassifier(n_estimators=n)
    model_n.fit(X_train,y_train)
    predictions = model_n.predict(X_test)
    err = 1 - accuracy_score(y_test,predictions)
    error_rate.append(err)
```

```python
plt.plot(range(1,96) ,error_rate)
```

#### From this we can see that after considering only n_estimators = 18(i.e. stumps) error rate got stabilized
