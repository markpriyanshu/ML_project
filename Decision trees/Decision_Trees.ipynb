```python
X = pd.get_dummies(df.drop('species' , axis = 1), drop_first=True)
```

```python
X
```

```python
y = df['species']
```

### No need to do feature scaling in Desicion tree

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_model = DecisionTreeClassifier()
```

```python
base_model.fit(X_train,y_train)
```

```python
base_y_pred = base_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix
```

```python
print(classification_report(y_test,base_y_pred))
```

```python
plot_confusion_matrix(base_model , X_test , y_test)
```

## Plotting Desicion Tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (12,8),dpi =200)
plot_tree(base_model , feature_names=base_model.feature_names_in_ , filled=True );
```

In above figure , ROOT NODE = 'flipper_length_mm' in which samples = 234 i.e. at root node the tree is getting 234 rows of data which is eqaul to len(X_train)

## Create a function which takes in different model and plot tree

```python
def report_model(model_name):
    model_preds = model_name.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize = (12,8),dpi =200)
    plot_tree(model_name , feature_names=base_model.feature_names_in_ , filled=True );
```

```python
model_entropy = DecisionTreeClassifier(criterion='entropy')
model_entropy.fit(X_train,y_train)
```

```python
report_model(model_entropy)
```

## Fixing maximum depth of tree

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
report_model(pruned_tree)
```
