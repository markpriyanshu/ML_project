```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales' , axis = 1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test= scaler.transform(X_test)
```

## ElasticNet

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_net_model = ElasticNet()
```

```python
parameters_grid = {'alpha' : [0.1,1,5,10,50,100] , 'l1_ratio' : [0.1 , 0.5 , 0.7 , 0.95 , 0.99 , 1]}
```

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model , param_grid=parameters_grid , 
                           scoring='neg_mean_squared_error' , cv = 5 , verbose=0)

# verbose = it gives us information of work going on.
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_model.cv_results_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test , y_pred)
```
